```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score

# Chargement des données RFM 
rfm= pd.read_csv("RFM_segmenté.csv", encoding="ISO-8859-1")
rfm
scaler = StandardScaler()
X_scaled = scaler.fit_transform(rfm[['Recence', 'Frequence', 'Montant']])
```

## Détermination du nombre optimal de clusters (méthode du coude)

```python
inertia = []
K_range = range(1, 10)

for k in K_range:
    kmeans = KMeans(n_clusters=k, n_init=10, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.plot(K_range, inertia, marker='o')
plt.title("Méthode du coude")
plt.xlabel("Nombre de clusters")
plt.ylabel("Inertie")
plt.grid(True)
plt.show()
```

## Application du K-means avec k=4

```python
kmeans = KMeans(n_clusters=4, n_init=10, random_state=42)
rfm['Cluster'] = kmeans.fit_predict(X_scaled)
```

## Visualisation des clusters

```python
plt.figure(figsize=(8,6))
sns.scatterplot(data=rfm, x='Recence', y='Montant', hue='Cluster', palette='Set2')
plt.title("Segmentation des clients (K-means)")
plt.xlabel("Recence")
plt.ylabel("Montant")
plt.legend(title="Cluster")
plt.show()
```

 ## Évaluation de la stabilité du modèle (scores silhouette pour différents seeds)

```python
stability_scores = []
for seed in range(10, 110, 10):
    model = KMeans(n_clusters=4, n_init=10, random_state=seed)
    labels = model.fit_predict(X_scaled)
    score = silhouette_score(X_scaled, labels)
    stability_scores.append((seed, score))

# Affichage des scores silhouette
for seed, score in stability_scores:
    print(f"Seed {seed} → Score silhouette : {score:.3f}")

```

## Profils clients par cluster

```python
summary = rfm.groupby('Cluster').agg({
    'Recence': 'mean',
    'Frequence': 'mean',
    'Montant': 'mean',
    'CustomerID': 'count'
}).rename(columns={'CustomerID': 'Nb_clients'}).sort_values('Recence')

summary.index.name = "Cluster"
summary.reset_index(inplace=True)
print(summary)
```

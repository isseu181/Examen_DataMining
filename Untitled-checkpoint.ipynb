```python
import pandas as pd
import numpy as np
from datetime import datetime
from sklearn.cluster import KMeans
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
import seaborn as sns

# Chargement des données
df = pd.read_csv("ecommerce_nettoye.csv", encoding="ISO-8859-1")


# Ajout de la colonne montant total
df['Mont_total'] = df['Quantity'] * df['UnitPrice']

# Conversion de la date
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])



# Calcul des données RFM
snapshot_date = df['InvoiceDate'].max() + pd.Timedelta(days=1)

rfm = df.groupby('CustomerID').agg({
    'InvoiceDate': lambda x: (snapshot_date - x.max()).days,   # Recency
    'InvoiceNo': 'nunique',                                     # Frequency
    'Mont_total': 'sum'                                         # Monetary
}).reset_index()

rfm.columns = ['CustomerID', 'Recency', 'Frequency', 'Monetary']

# Normalisation
scaler = StandardScaler()
rfm_scaled = scaler.fit_transform(rfm[['Recency', 'Frequency', 'Monetary']])

# Test de stabilité à l'initialisation (différents random_state)
results = []
for seed in range(5):
    kmeans = KMeans(n_clusters=10, random_state=seed, n_init=10)
    clusters = kmeans.fit_predict(rfm_scaled)
    inertia = kmeans.inertia_
    results.append((seed, inertia))

print("Inerties selon l'initialisation :")
for seed, inertia in results:
    print(f"Random State {seed}: Inertia = {inertia:.2f}")

# Appliquer K-Means avec random_state stable
kmeans = KMeans(n_clusters=10, random_state=0, n_init=10)
rfm['Cluster'] = kmeans.fit_predict(rfm_scaled)

# Résumé par cluster
cluster_summary = rfm.groupby('Cluster').agg({
    'Recency': 'mean',
    'Frequency': 'mean',
    'Monetary': ['mean', 'count']
}).round(2)

print("\nRésumé des segments clients (RFM):")
print(cluster_summary)

# Visualisation
plt.figure(figsize=(10, 6))
sns.scatterplot(data=rfm, x='Recency', y='Monetary', hue='Cluster', palette='tab10')
plt.title("Segmentation des clients par K-Means")
plt.show()

#  Proposition de contrat de maintenance par segment
contrat = {
    0: "Contrat standard",
    1: "Contrat premium",
    2: "Contrat de fidélisation",
    3: "Relance client",
    4: "Contrat découverte",
    5: "Contrat VIP",
    6: "Contrat occasionnel",
    7: "Relance promotionnelle",
    8: "Suivi personnalisé",
    9: "Contrat d'essai"
}

rfm['Contrat'] = rfm['Cluster'].map(contrat)

# Exporter si besoin
rfm.to_csv("clients_segmentes.csv", index=False)

print("\nContrat proposé par client (extrait) :")
print(rfm[['CustomerID', 'Cluster', 'Contrat']].head())
```

```python
# Calcul des quartiles
quantiles = rfm[['Recency', 'Frequency', 'Monetary']].quantile([0.25, 0.5, 0.75])
quantiles = quantiles.to_dict()

# Fonctions de scoring
def r_score(x):
    if x <= quantiles['Recency'][0.25]:
        return 4
    elif x <= quantiles['Recency'][0.5]:
        return 3
    elif x <= quantiles['Recency'][0.75]:
        return 2
    else:
        return 1

def fm_score(x, var):
    if x <= quantiles[var][0.25]:
        return 1
    elif x <= quantiles[var][0.5]:
        return 2
    elif x <= quantiles[var][0.75]:
        return 3
    else:
        return 4

# Application des scores
rfm['R'] = rfm['Recency'].apply(r_score)
rfm['F'] = rfm['Frequency'].apply(lambda x: fm_score(x, 'Frequency'))
rfm['M'] = rfm['Monetary'].apply(lambda x: fm_score(x, 'Monetary'))
rfm['RFM_Score'] = rfm['R'].map(str) + rfm['F'].map(str) + rfm['M'].map(str)

# Concatenation des scores RFM
rfm['RFM_concat_score']=rfm['R'].map(str) + rfm['F'].map(str) +rfm['M'].map(str)
rfm.head(10)

```
